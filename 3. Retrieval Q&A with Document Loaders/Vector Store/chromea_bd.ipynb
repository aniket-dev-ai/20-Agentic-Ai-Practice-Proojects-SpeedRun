```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_ollama import OllamaEmbeddings

embeddings = OllamaEmbeddings(
    model="qwen3-embedding:0.6b",
    dimensions=1024
)
```

```python
text_splitter = SemanticChunker(
    embeddings=embeddings,
    breakpoint_threshold_type="percentile",
    breakpoint_threshold_amount=70
)
```

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader("../Document Loader/Example.pdf")

doc = loader.load()
```

```python
doc
```

```python
for i, d in enumerate(doc):
    print(f"len of doc number {i}")
    print(len(d.page_content))
    print(f"metadata of doc number {i}")
    print(len(d.metadata))
    print("\n---\n")
```

```python
chunks = text_splitter.split_documents(doc)
```

```python
chunks
```

```python
for i, d in enumerate(chunks):
    print(f"len of chunk number {i}")
    print(len(d.page_content))
    print(f"metadata of chunk number {i}")
    print(len(d.metadata))
    print("\n---\n")
```

```python
vector_store = Chroma(
    embedding_function=embeddings,
    persist_directory="chroma_db",
    collection_name="example_collection"
)
```

```python
vector_store.add_documents(chunks)
```

```python
ques = [
    {"question": "What are the differences between UVA, UVB, and UVC radiation?"},
    {"question": "How does UV radiation damage human skin?"},
    {"question": "What is Fitzpatrick skin type classification and why is it used?"},
    {"question": "What physical methods can protect skin from sunlight besides sunscreen?"},
    {"question": "What are the main types of sunscreens based on mode of application?"},
    {"question": "Which organic UVB, UVA, and broad-spectrum filters are listed in the review?"},
    {"question": "Which inorganic sunscreen agents are described?"},
    {"question": "What natural chemicals are mentioned as photoprotective agents?"},
    {"question": "How do sunscreens work according to the review?"},
    {"question": "What are the sunscreen regulations and labeling rules in different countries?"},
    {"question": "What is SPF and how is it defined?"},
    {"question": "What is the critical wavelength method and how is broad-spectrum protection determined?"},
    {"question": "What are the in-vitro methods used to evaluate sunscreen effectiveness?"},
    {"question": "What are IPD, PPD, PFA, and PPF, and how do they differ?"},
    {"question": "How is sunscreen photostability measured using AUCI?"},
    {"question": "What is the recommended sunscreen application dose per square centimeter?"},
    {"question": "How often should sunscreen be reapplied?"},
    {"question": "What are the controversies and limitations of sunscreen use?"},
    {"question": "Which natural compounds like flavonoids, tannins, carotenoids, and vitamins are discussed for photoprotection?"},
    {"question": "What recent sunscreen technologies and product forms are mentioned?"}
]
```

```python
for q in ques:  # Limit to first 5 questions
    print(f"Question: {q['question']}")
    result = vector_store.similarity_search(q["question"], k=2)
    for i, r in enumerate(result):
        print(f"Result {i+1}:")
        print(r.page_content)
        print(f"Metadata: {r.metadata}")
        print("\n---\n")
```
