```python
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
load_dotenv()

model = ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite", temperature=0.3)
```

```python
template1 = PromptTemplate(
    template="Write a detailed report about this topic: {topic}",
    input_variables=["topic"]
)
```

```python
template2 = PromptTemplate(
    template="Write a brief summary on the following \ntext: {text}",
    input_variables=["text"]
)
```

```python
prompt1 = template1.format(topic="Artificial Intelligence")
```

```python
result = model.invoke(prompt1)
```

```python
result
```

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
```

```python
from IPython.display import Markdown , display

display(Markdown(result.content[0]['text']))
```

```python
chain = template1 | model | parser | template2 | model | parser
```

```python
res = chain.invoke({"topic": "Artificial Intelligence"})
```

```python
display(Markdown(res))
```
